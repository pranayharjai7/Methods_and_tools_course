````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline   # display plots inline in Jupyter
````

## 1. Line Plots — `plt.plot`

Used for: loss/cost curves over training epochs.

````python
# Simulate a training loss curve
epochs = np.arange(100)
loss = 2.0 * np.exp(-0.05 * epochs) + 0.1 * np.random.randn(100) * np.exp(-0.05 * epochs)

plt.figure(figsize=(8, 4))
plt.plot(epochs, loss, 'b-', linewidth=2, label='Training loss')
plt.xlabel('Epoch')
plt.ylabel('Cost J(W)')
plt.title('Gradient Descent Convergence')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Multiple lines
x = np.linspace(-5, 5, 200)

plt.figure(figsize=(8, 4))
plt.plot(x, 1 / (1 + np.exp(-x)),  'b-', label='sigmoid(x)')
plt.plot(x, np.maximum(0, x),       'r--', label='ReLU(x)')
plt.plot(x, np.tanh(x),             'g:',  label='tanh(x)')
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlabel('z')
plt.ylabel('Activation')
plt.title('Common Activation Functions')
plt.legend()
plt.ylim(-1.5, 1.5)
plt.grid(True, alpha=0.3)
plt.show()
````

## 2. Scatter Plots — `plt.scatter`

Used for: visualizing data points colored by class label.

````python
# Generate 2-class data
np.random.seed(42)
X_pos = np.random.randn(50, 2) + [2, 2]
X_neg = np.random.randn(50, 2) + [-2, -2]

plt.figure(figsize=(6, 5))
plt.scatter(X_pos[:, 0], X_pos[:, 1], c='blue', marker='+', s=60, label='y=1')
plt.scatter(X_neg[:, 0], X_neg[:, 1], c='red',  marker='o', s=40, label='y=0')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data')
plt.legend()
plt.show()
````

````python
# Plot using a color array — useful when you have a label array
np.random.seed(0)
X_all = np.vstack([X_pos, X_neg])
y_all = np.array([1]*50 + [0]*50)

plt.figure(figsize=(6, 5))
plt.scatter(X_all[:, 0], X_all[:, 1], c=y_all, cmap='bwr', edgecolors='k', s=50)
plt.colorbar(label='Class')
plt.title('Data colored by label')
plt.show()
````

## 3. Decision Boundary

How the exam notebooks plot the logistic regression boundary line.

````python
# For linear decision boundary: W[0] + W[1]*x1 + W[2]*x2 = 0
# Solve for x2: x2 = -(W[0] + W[1]*x1) / W[2]

W = np.array([-0.5, 1.0, 1.0])  # example weights [bias, w1, w2]

x1_range = np.linspace(-4, 4, 100)
x2_boundary = -(W[0] + W[1] * x1_range) / W[2]

plt.figure(figsize=(6, 5))
plt.scatter(X_pos[:, 0], X_pos[:, 1], c='blue', marker='+', s=60, label='y=1')
plt.scatter(X_neg[:, 0], X_neg[:, 1], c='red',  marker='o', s=40, label='y=0')
plt.plot(x1_range, x2_boundary, 'k-', linewidth=2, label='Decision boundary')
plt.xlabel('x1'); plt.ylabel('x2')
plt.title('Logistic Regression Decision Boundary')
plt.legend()
plt.ylim(-6, 8)
plt.show()
````

## 4. Subplots — `plt.subplots`

Used for: side-by-side comparisons, showing multiple training curves.

````python
epochs = np.arange(200)
train_loss = 1.5 * np.exp(-0.03 * epochs) + 0.05
val_loss   = 1.5 * np.exp(-0.025 * epochs) + 0.15
accuracy   = 1 - np.exp(-0.025 * epochs)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left subplot
axes[0].plot(epochs, train_loss, 'b-', label='Train')
axes[0].plot(epochs, val_loss,   'r--', label='Validation')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training & Validation Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Right subplot
axes[1].plot(epochs, accuracy, 'g-', linewidth=2)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training Accuracy')
axes[1].set_ylim(0, 1)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 5. Displaying Images — `plt.imshow`

Used for: MNIST digit visualization, K-means image compression.

````python
# Display a single 28x28 grayscale image
np.random.seed(5)
fake_digit = np.random.rand(28, 28)

plt.figure(figsize=(3, 3))
plt.imshow(fake_digit, cmap='gray')
plt.axis('off')   # hide tick marks
plt.title('Digit (fake)')
plt.show()
````

````python
# Display a grid of images — pattern from neural network notebooks
n_images = 25
images = np.random.rand(n_images, 28, 28)
labels = np.random.randint(0, 10, n_images)

fig, axes = plt.subplots(5, 5, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')
    ax.set_title(str(labels[i]), fontsize=10)
    ax.axis('off')

plt.suptitle('Sample Digits', y=1.02)
plt.tight_layout()
plt.show()
````

## 6. Contour Plots

Used for: visualizing cost function surfaces in gradient descent notebooks.

````python
# Create a 2D grid of parameter values
w0_vals = np.linspace(-5, 5, 100)
w1_vals = np.linspace(-5, 5, 100)
W0, W1 = np.meshgrid(w0_vals, w1_vals)

# Compute a bowl-shaped cost function J(w0, w1)
J = W0**2 + 2 * W1**2 + W0 * W1   # example quadratic

plt.figure(figsize=(7, 5))
contours = plt.contour(W0, W1, J, levels=20, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)
plt.xlabel('w0 (bias)')
plt.ylabel('w1')
plt.title('Cost Function Contours')

# Mark the minimum
plt.scatter(0, 0, color='red', s=100, zorder=5, label='Minimum')
plt.legend()
plt.colorbar(label='J(w0, w1)')
plt.show()
````

## 7. Plot Cluster Assignments (K-Means)

````python
# Generate 3-cluster data
np.random.seed(0)
c1 = np.random.randn(50, 2) + [0, 0]
c2 = np.random.randn(50, 2) + [5, 5]
c3 = np.random.randn(50, 2) + [0, 8]
X_clust = np.vstack([c1, c2, c3])

# Fake assignments and centroids
assignments = np.array([0]*50 + [1]*50 + [2]*50)
centroids = np.array([[0.1, 0.2], [4.8, 5.1], [0.0, 7.9]])

colors = ['red', 'blue', 'green']
plt.figure(figsize=(6, 6))
for k in range(3):
    mask = assignments == k
    plt.scatter(X_clust[mask, 0], X_clust[mask, 1],
                c=colors[k], alpha=0.6, label=f'Cluster {k}')

# Mark centroids
plt.scatter(centroids[:, 0], centroids[:, 1],
            c='black', marker='X', s=200, zorder=5, label='Centroids')

plt.xlabel('x1'); plt.ylabel('x2')
plt.title('K-Means Clustering Result')
plt.legend()
plt.show()
````

### Exercise 4: Cost Surface
Plot the contour lines for the MSE cost `J(w0, w1) = mean((w0 + w1*x - y)^2)` over a grid of `w0` and `w1` values, given the data below.

````python
np.random.seed(42)
x_data = np.random.randn(30)
y_data = 2.0 * x_data + 1.0 + 0.3 * np.random.randn(30)  # true: w0=1, w1=2

w0_vals = np.linspace(-2, 4, 80)
w1_vals = np.linspace(-1, 5, 80)
W0, W1 = np.meshgrid(w0_vals, w1_vals)

# YOUR CODE HERE: compute J at each (w0, w1) point and plot contours
# Hint: J[i,j] = mean((W0[i,j] + W1[i,j]*x_data - y_data)^2)
````

````python
# SOLUTION
J = np.zeros_like(W0)
for i in range(W0.shape[0]):
    for j in range(W0.shape[1]):
        pred = W0[i,j] + W1[i,j] * x_data
        J[i,j] = np.mean((pred - y_data)**2)

plt.figure(figsize=(7, 5))
plt.contour(W0, W1, J, levels=30, cmap='viridis')
plt.scatter([1.0], [2.0], c='red', s=100, zorder=5, label='True (w0=1, w1=2)')
plt.xlabel('w0 (bias)'); plt.ylabel('w1')
plt.title('MSE Cost Surface')
plt.colorbar(label='Cost J')
plt.legend()
plt.show()
````
